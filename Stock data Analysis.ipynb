```python
df_dividend = data.dividends
```

```python
df_holders = data.institutional_holders
```

```python
data = df_dividend.resample('YE').sum().reset_index()
```

```python
print(data.head())
```

```python
data['Year']= data['Date'].dt.year
data[:5]
```

Visualizing the Data

```python
plt.figure(figsize=(10,6))
plt.bar(data['Year'], data['Dividends'])
plt.title('Microsoft Dividends History')
plt.xlabel('Year')
plt.ylabel('Dividend Yield($)')

plt.show()
```

Microsoft paid a larger amount of dividends in 2004

In July 2004, Microsoft's board of directors approved a plan that included:

A special one-time dividend of $3.00 per share, totaling $32 billion.
An increase in the company's regular annual dividend.
A share repurchase program to buy back up to $30 billion of Microsoft stock over the next four years.
This strategic move was intended to enhance shareholder value by distributing excess cash, improving the company's financial structure, and demonstrating confidence in its ongoing profitability and cash generation capabilities.

```python
plt.figure(figsize=(10,6))
plt.bar(data['Year'], data['Dividends'])
plt.title('Microsoft Dividends History')
plt.xlabel('Year')
plt.ylabel('Dividend Yield($)')
plt.xlim(2002,2023)
plt.show()
```

Microsoft increased her dividends consistently over the last 21 years. Inflexion occured in 2010

By 2010, Microsoft faced increasing competition from companies like Apple and Google in various segments, including operating systems, mobile devices, and online services. The company’s decision to increase dividends could also be seen as a way to maintain investor interest and confidence in the face of growing competitive pressures.

The inflection point in Microsoft's dividends in 2010 was driven by its strong financial position, a strategic shift towards returning capital to shareholders, and the broader economic recovery. This decision reflected Microsoft's maturity as a company and its commitment to providing consistent returns to its investors.

```python
data= df_holders.groupby('Holder')['Value'].sum().reset_index()
data
```

```python
plt.figure(figsize=(10,6))
plt.bar(data['Holder'], data['Value'])
plt.title('Microsoft Shareholders by Value')
plt.xlabel('Holders')
plt.xticks(rotation=90)
plt.ylabel('Values')
plt.show()
```

Vanguard Group Inc is the largest holder of Microsoft shares followed by Blackrock Inc.
While Norges Bank Investment Management is the least.

```python
data= df_holders.groupby('Holder')['pctHeld'].sum()
data
```

```python
explode_list = [0.1] * 10
data.plot(kind='pie',
                            figsize=(15, 6),
                            autopct='%1.1f%%', 
                            startangle=90,    
                            shadow=True,       
                            labels=None,      
                            pctdistance=1.12,   
                            explode=explode_list 
                            )
plt.title('Holders by Percentage', y=1.12, fontsize = 15) 
plt.axis('equal') 
plt.legend(labels=data.index, loc='upper left', fontsize=10) 

plt.show()
```
